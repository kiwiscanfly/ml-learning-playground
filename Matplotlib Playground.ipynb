```python
%pip install matplotlib numpy

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#Random Data
years = [2010 + x for x in range(10)]
values = np.random.rand(10)
```

## Line Graph

```python
plt.plot(years, values, c="#A49FE0", lw=4, linestyle='--', marker='o', markersize=8, markerfacecolor='#D09BAF', markeredgewidth=2, markeredgecolor='#A16A6E')
plt.title('Hello World')

# Save to disk
plt.savefig('plot.png')

plt.show()
```

```python
plt.plot(years, values, label='Random')

fontdict = {'fontname': 'Comic Sans MS', 'fontsize': 16}

plt.xlabel('Year', fontdict=fontdict)
plt.ylabel('Random Value', fontdict=fontdict)
plt.title('Randoms Over Time', fontdict=fontdict)

plt.xticks(years)
plt.yticks(np.linspace(0, 1, 5))

plt.legend()
plt.show() 
```

```python
# Shorthand notation
plt.plot(years, values, 'r--', label='Random')

# Line 2
values2 = np.random.rand(30)
years2 = [2010 + x/3 for x in range(30)]
plt.plot(years2, values2, 'b-', label='Random 2')

plt.show()
```

## Scatter Plot

```python
x_data = np.random.random(60) * 100
y_data = np.random.random(60) * 100

plt.scatter(x_data, y_data, marker="*", c="#D09BAF", s=150, alpha=0.5)

plt.show()
```

## Bar Chart

```python
# Bar Chart
x = np.arange(10)
y = np.random.random(10) * 100

plt.bar(x, y, color='#A49FE0')
```

# With Pandas

```python
gas = pd.read_csv('gas_prices.csv')

plt.title('Gas Prices Over Time (in USD)')

usa = gas[gas['Country'] == 'USA']
canada = gas[gas['Country'] == 'Canada']
germany = gas[gas['Country'] == 'Germany']

plt.plot(usa['Year'], usa['Price'], 'b.-', label='USA')
plt.plot(canada['Year'], canada['Price'], 'r.-', label='Canada')
plt.plot(germany['Year'], germany['Price'], 'g.-', label='Germany')

plt.xticks(gas['Year'][::3].tolist() + [2011])

plt.xlabel('Year')
plt.ylabel('Price in USD')
plt.legend()
plt.show()
```
